```python
## create a column age_group
## quantile binning concept 
labels = ['Young', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
##create purchase_frequency_days column
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days']= df['frequency_of_purchases'].map(frequency_mapping)
```

```python
from sqlalchemy import create_engine
import urllib.parse

# Step 1: Connection details
username = "postgres"                 # default user
password = urllib.parse.quote_plus("Password@098")  # encode special chars
host = "localhost"                    # NOT 5432
port = 5432                           # integer, not string
database = "customer_behaviour"

# Step 2: Create engine
engine = create_engine(
    f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"
)

# Step 3: Load dataframe to PostgreSQL
table_name = "customer"
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```

```python
from sqlalchemy import text

table_to_drop = "customertable"

with engine.begin() as conn:
    conn.execute(text(f'DROP TABLE IF EXISTS "{table_to_drop}" CASCADE;'))

print("Table dropped successfully!")
```
